#  Analyzing Data with Pandas and Visualizing Results with Matplotlib

**Objectives**
- Load and explore the Iris dataset using pandas
- Perform basic analysis (stats, groupby)
- Create 4 visualizations (line, bar, histogram, scatter)
- Present findings


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris(as_frame=True)
df = iris.frame
```

```python
# Show first 5 rows
df.head()
```

```python
# Data types
print(df.dtypes)

# Missing values
print(df.isnull().sum())
```

```python
# Clean data (just in case)
df = df.dropna()
```

```python
# Summary statistics
df.describe()
```

```python
# Group by target (species)
grouped_means = df.groupby("target").mean()
grouped_means
```

```python
# Line chart
plt.figure(figsize=(8,5))
plt.plot(df.index, df["sepal length (cm)"], label="Sepal Length")
plt.title("Line Chart: Sepal Length Across Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.show()
```

```python
# Bar chart
plt.figure(figsize=(8,5))
species_names = iris.target_names
plt.bar(species_names, grouped_means["petal length (cm)"])
plt.title("Bar Chart: Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length (cm)")
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8,5))
plt.hist(df["sepal width (cm)"], bins=15)
plt.title("Histogram: Sepal Width Distribution")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8,5))
plt.scatter(df["sepal length (cm)"], df["petal length (cm)"], c=df["target"], cmap="viridis")
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.colorbar(label="Species (0=setosa, 1=versicolor, 2=virginica)")
plt.show()
```

##  Observations / Findings

1. **Setosa** has much smaller petal lengths than the other species.  
2. **Sepal width distribution** is roughly normal.  
3. **Virginica** shows the largest average petal length.  
4. The **scatter plot** shows clear separation between *Setosa* and the other species.
